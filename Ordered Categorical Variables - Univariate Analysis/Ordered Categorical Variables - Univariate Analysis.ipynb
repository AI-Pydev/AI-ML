## What is the mode of the num_keywords variable in the News Popularity data set?

```python
df1 = pd.read_csv('popularity.csv')
```

```python
df1.shares.describe()
```

```python
df1.shares.describe(percentiles = [0.78, 0.80])
```

```python
df1 = df1.drop('url', axis=1)
```

```python
low = .05
high = .95
quant_df = df1.quantile([low, high])
print(quant_df)
```

```python
df2 = df1.apply(lambda x: x[(x>quant_df.loc[low,x.name]) & 
                                    (x < quant_df.loc[high,x.name])])
```

```python
df2.describe()
```
