### Dummy Variables

```python
# Use pandas library to create the dummy variables
dummies = pd.get_dummies(categorical_data, drop_first=True)
dummies.head()
```

### Building model using statsmodel, for the detailed statistics

```python
# Import the statsmodels 
import statsmodels.api as sm  
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Creaed below methods to avoid writing the code multiple times and to test the models

# Method to build model
def build_stats_model(X,y):
    X = sm.add_constant(X) # Adding a constant variable 
    lm = sm.OLS(y,X).fit() # Running the linear model
    print(lm.summary()) #Let's see the summary of our linear model
    return X

# Method to check VIF value
def check_vif(X):
    vif = pd.DataFrame()
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return(vif)
```
